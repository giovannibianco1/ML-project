```python

# Calculate the lower and upper bounds for outliers
Q1 = df_no_na.quantile(0.25)  # First quartile
Q3 = df_no_na.quantile(0.75)  # Third quartile
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers based on the lower and upper bounds
outliers = df_no_na[(df_no_na < lower_bound) | (df_no_na > upper_bound)]

# Filter out the outliers
df_no_na = df_no_na[~df_no_na.isin(outliers)]
```

```python


# Assuming df_no_na is your DataFrame
import matplotlib.pyplot as plt

# Assuming df_no_na is your DataFrame
ok = df_no_na[df_no_na['SeriousDlqin2yrs'] == 1]
no = df_no_na[df_no_na['SeriousDlqin2yrs'] == 0]

# Create a figure and axes
num_vars = df_no_na.columns[df_no_na.columns!='SeriousDlqin2yrs']

fig, axes = plt.subplots(nrows=len(num_vars), figsize=(10, 5 * len(num_vars)))

# Plot histograms for each column
for ax, var in zip(axes, num_vars):
    bins = np.histogram(np.hstack((ok[var], no[var])), bins=100)[1]  # Get shared bin edges
    range=(min(df_no_na[var]),max(df_no_na[var]))
    print(range)
    ax.hist(ok[var], bins=bins, color='red', alpha=0.5, label='SeriousDlqin2yrs = 1',density=True,range=range)
    ax.hist(no[var], bins=bins, color='blue', alpha=0.5, label='SeriousDlqin2yrs = 0',density=True,range=range)
    ax.set_title(var)
    ax.legend()

# Set tight layout and show plot
plt.tight_layout()
plt.show()
```
